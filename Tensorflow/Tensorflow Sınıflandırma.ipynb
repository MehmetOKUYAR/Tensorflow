```python
from sklearn.model_selection import train_test_split #ikiye ayrımak için kullanacağımız modülü import ettik
```

```python
# y=wx+b
# y-->label
y = dataFrame["Type"].values 

# x-->feature(özellik)
x = dataFrame.drop("Type",axis=1).values #geri kalan kolonların hepsi x olacak

#veriyi burada ikiye böldük yüzde 33'ünü test versisi geri kalanını öğrenme verisi olarak ayırdık
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15) 
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_train
```

```python
x_test =scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
# burada modelimizi belirledik
model = Sequential() 

model.add(Dense(30,activation="relu")) #gizli katmanlarımızı ekliyoruz
model.add(Dense(15,activation="relu")) #ve içinde kaç nöron olacağını belirdik
model.add(Dense(15,activation="relu")) #ve son olarak aktivasyon modelimizi seçtik
model.add(Dense(1,activation="sigmoid")) #ve son olarak çıktı modelimizi seçtik



model.compile(optimizer="adam",loss ="binary_crossentropy") # gradient descent (optimizasyon) algoritmamızı belirledik.
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=700,verbose=1) #burada verilerimizi eğitmeye başlıyoruz
```

```python
model.history.history
```

```python
modelkaybi = pd.DataFrame(model.history.history)
```

```python
modelkaybi.plot()
```

```python
earlyStopping = EarlyStopping("val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=700,verbose=1,callbacks=[earlyStopping]) #burada verilerimizi eğitmeye başlıyoruz
```

```python
modelkaybi = pd.DataFrame(model.history.history)
```

```python
modelkaybi.plot()
```

```python
# burada modelimizi belirledik
model = Sequential() 

model.add(Dense(30,activation="relu")) #gizli katmanlarımızı ekliyoruz
model.add(Dropout(0.5))

model.add(Dense(15,activation="relu")) #ve içinde kaç nöron olacağını belirdik
model.add(Dropout(0.5))

model.add(Dense(15,activation="relu")) #ve son olarak aktivasyon modelimizi seçtik
model.add(Dropout(0.5))

model.add(Dense(1,activation="sigmoid")) #ve son olarak çıktı modelimizi seçtik



model.compile(optimizer="adam",loss ="binary_crossentropy") # gradient descent (optimizasyon) algoritmamızı belirledik
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=700,verbose=1,callbacks=[earlyStopping]) #burada verilerimizi eğitmeye başlıyoruz
```

```python
kayipDF = pd.DataFrame(model.history.history)
```

```python
kayipDF.plot()
```

```python
tahmin = model.predict_classes(x_test)
```

```python
tahmin
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```
